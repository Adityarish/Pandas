## Importing Pandas

```python
import pandas as pd
import numpy as np
```

```python
# string
country = ['India', 'Nepal', 'America','Russian', 'Afganistan']
pd.Series(country) # all data are of object type
```

```python
# integers

runs = [13, 24, 25, 876,25,21]
pd.Series(runs, dtype = np.int16)
```

```python
# custom index

marks = [67, 76, 89, 42, 98]
subjects  = ['Maths', 'English', 'Science', 'DSA', 'SSt']

pd.Series(marks, index = subjects, name = "Arushi ka marks")
```

## Series from dictionary

```python
marks = {
    'maths' : 56,
    'english' : 78,
    'science' : 89,
    'hindi' : 100

}

marks_series = pd.Series(marks, name = "arushi ka marks")
marks_series
```
